```python
#$$$$$$ex5-2
f1=open(r'd:\data\score.txt')
head=f1.readline()
import os
import re
os.chdir(r'd:\data')      #当指定工作路径之后就不用再说明文件位置
with open('level.txt','w') as f2:     ####使用with open可以避免忘记 close
    for row in f1:
        line=re.split('\t',row)
        lsn=list(map(lambda x:int(x),line))
        sno=lsn[0]
        ssum=lsn[1]+lsn[2]+lsn[3]
        if ssum>=260:
            grade='优秀'
        elif ssum>=180:
            grade='及格'
        else:
            grade='不及格'
        f2.write(str(sno)+'\t'+grade+'\n')
    f1.close
```

```python
def segged_s(s):
    return ''.join(jieba.lcut(s))
df3=finnews['segged_text']=finnews['text'].apply(segged_s)
print(df3)
```

```python
seg_list = jieba.cut("我来到北京清华大学", cut_all=True)
print("全模式: " + "/ ".join(seg_list))  # 全模式

seg_list = jieba.cut("我来到北京清华大学", cut_all=False)
print("精准模式: " + "/ ".join(seg_list))  # 精确模式

seg_list = jieba.cut("他来到了网易杭研大厦")  # 默认是精确模式
print(", ".join(seg_list))
```

```python
###########$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$如何把这个数据按某一行元素进行降序或者升序排列

import jieba
def segged_s_len(s):
    return len(jieba.lcut(s))
finnews['Nwords']=finnews['text'].apply(segged_s_len)
finnews.sort_values(by='Nwords',ascending=False)
```
